## Data Visualization

```python
def visualize(feature):
        mdr = df1[df1.Class=='MDR-TB']
        mdr = mdr[feature]
        tb = df1[df1.Class=='TB']
        tb = tb[feature]

        plt.xlabel(f"{feature}")
        plt.ylabel("Number Of Cases")
        plt.title("Multi-Drug Resistant Tuberculosis Visualization")


        plt.hist([mdr, tb], rwidth=0.95, color=['green','red'],label=['MDR-TB','TB'])
        plt.legend()
```

```python
visualize('Gender')
```

```python
visualize('Alcohol')
```

```python
visualize('Age')
```

## Let's print unique values in object columns to see data values

```python
def print_unique_col_values(df):
       for column in df:
            if df[column].dtypes=='object':
                print(f'{column}: {df[column].unique()}') 
```

```python
print_unique_col_values(df1)
```

## Convert Yes and No column values to 1 or 0

```python
yes_no_columns = ['Smoking','Alcohol','Cavitary pulmonary','Diabetes','Tboutside the lungs']
for col in yes_no_columns:
    df1[col].replace({'Yes': 1,'No': 0},inplace=True)
```

```python
for col in df1:
    print(f'{col}: {df1[col].unique()}') 
```
